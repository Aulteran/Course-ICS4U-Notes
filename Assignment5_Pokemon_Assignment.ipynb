```python
# Part A - Formatting the Data: (All changes should be made to the existing (original) dataframe)
    # 1. Format the Database so it shows a maximum of two decimal places
    # 2. Format the Database so it shows all columns
    # 3. Display the number of rows and columns in the Dataframe
    # 4. Change the ‘Generation’ Column name to ‘Gen’
    # 5. Set a default value of ‘NA’ for all missing cells in the Database
    # 6. Change the dType of the TYPE 2 series to Category
    # 7. Display the Datatype of each of the series in the Dataframe (hint: info)
    # 8. Create a NEW Dataframe with 2 Pokemon you Make up: Add this new Dataframe and the existing Pokemon Dataframe together to create your new working Dataframe all_pokemon
    # 9. Display the last 10 pokemon of of the all_pokemon dataframe 
```

```python
# PartA
pd.set_option('display.float_format', '{:.2f}'.format)
pd.set_option('display.max_columns', None)
print("Number of rows:", pokemon_data.shape[0])
print("Number of columns:", pokemon_data.shape[1])
pokemon_data.rename(columns={'Generation': 'Gen'}, inplace=True)
pokemon_data.fillna('NA', inplace=True)
pokemon_data['Type 2'] = pokemon_data['Type 2'].astype('category')
pokemon_data.info()

# uh question 8 is kinda weird oop
# Create a new DataFrame with two Pokemon
new_pokemon_data = pd.DataFrame({
    'NAME': ['Fictimon1', 'Fictimon2'],
    'Type 1': ['Fairy', 'Dragon'],
    'Type 2': ['None', 'Fire'],
    'HP': [80, 90],
    'Attack': [100, 110],
    'Defense': [75, 85],
    'Sp. Atk': [120, 130],
    'Sp. Def': [80, 90],
    'Speed': [100, 110],
    'Generation': [8, 8],
    'Legendary': [False, False]
})
# Display the new DataFrame
print("\nnew_pokemon_data:")
print(new_pokemon_data)
# Concatenate the two DataFrames to create 'all_pokemon'
all_pokemon = pd.concat([pokemon_data, new_pokemon_data], ignore_index=True)

# q9
print(all_pokemon.tail(10))

all_pokemon.head()
```
