```python
'''
Author: Aadil Hussain
Built on: Python 3.10.11
'''

import pandas as pd

POKEDATA_FILEPATH = 'Assignment5_Pokemon_Data.csv'

pokemon_data = pd.read_csv(POKEDATA_FILEPATH)

# Download the Pokémon Dataset and create a Jupyter Notebook that displays the following answers to these questions
# (NOTE: create a DataFrame to answer these questions)
```

```python
# Part A - Formatting the Data: (All changes should be made to the existing (original) dataframe)
    # 1. Format the Database so it shows a maximum of two decimal places
    # 2. Format the Database so it shows all columns
    # 3. Display the number of rows and columns in the Dataframe
    # 4. Change the ‘Generation’ Column name to ‘Gen’
    # 5. Set a default value of ‘NA’ for all missing cells in the Database
    # 6. Change the dType of the TYPE 2 series to Category
    # 7. Display the Datatype of each of the series in the Dataframe (hint: info)
    # 8. Create a NEW Dataframe with 2 Pokemon you Make up: Add this new Dataframe and the existing Pokemon Dataframe together to create your new working Dataframe all_pokemon
    # 9. Display the last 10 pokemon of of the all_pokemon dataframe 
```

```python
# PartA
pd.set_option('display.float_format', '{:.2f}'.format)
pd.set_option('display.max_columns', None)
print("Number of rows:", pokemon_data.shape[0])
print("Number of columns:", pokemon_data.shape[1])
pokemon_data.rename(columns={'Generation': 'Gen'}, inplace=True)
pokemon_data.fillna('NA', inplace=True)
pokemon_data['Type 2'] = pokemon_data['Type 2'].astype('category')
pokemon_data.info()

# uh question 8 is kinda weird oop
# Create a new DataFrame with two Pokemon
new_pokemon_data = pd.DataFrame({
    'NAME': ['Fictimon1', 'Fictimon2'],
    'Type 1': ['Fairy', 'Dragon'],
    'Type 2': ['None', 'Fire'],
    'HP': [80, 90],
    'Attack': [100, 110],
    'Defense': [75, 85],
    'Sp. Atk': [120, 130],
    'Sp. Def': [80, 90],
    'Speed': [100, 110],
    'Gen': [8, 8],
    'Legendary': [False, False]
})
# Display the new DataFrame
print("\nnew_pokemon_data:")
print(new_pokemon_data)
# Concatenate the two DataFrames to create 'all_pokemon'
all_pokemon = pd.concat([pokemon_data, new_pokemon_data], ignore_index=True)

# q9
print(all_pokemon.tail(10))

all_pokemon.head()
```

```python
# Part C # 1.

# Create a bar plot for the distribution of Pokemon types (Type 1)
type_counts = all_pokemon['Type 1'].value_counts()
print(type_counts)
#pip install matplotlib?
type_counts.plot(kind='pie') #why is my plot not plotting?
```

```python
#PartC #2

# Filter Pokémon from the 4th generation with HP over 90
gen4_high_hp_pokemon = all_pokemon[(all_pokemon['Gen'] == 4) & (all_pokemon['HP'] > 90)]

# Display the overall number of Pokémon that meet the criteria
overall_count = len(gen4_high_hp_pokemon)
print(f"Overall number of 4th Gen Pokémon with HP over 90: {overall_count}")

# Display relevant information in a bar graph
gen4_high_hp_pokemon.plot(x='Name', y='HP', kind='bar', rot = 80)
```

```python
#PartC #3

# Filter the DataFrame for the First Generation (Gen 1)
first_gen_pokemon = all_pokemon[all_pokemon['Gen'] == 1]

# Calculate the Count of each Type 1 in the first generation
type_counts_first_gen = first_gen_pokemon['Type 1'].value_counts()

# Display the Count in a Bar Graph
type_counts_first_gen.plot(kind='bar', figsize=(12, 6), color='skyblue')
```

```python
#PartC #4

# Scatter plot of Attack vs. Defense
all_pokemon.plot.scatter(x='Attack', y='Defense', color='skyblue', figsize=(10, 6))
#spread gets wider as dataset approaches higher values, but imagining a trendline supports the data
```

```python
#PartC #5 - Visualize the data between Legendary and Non-Legendary Pokemon

# Calculate the count of Legendary and Non-Legendary Pokémon
legendary_counts = all_pokemon['Legendary'].value_counts()

# graph the counts
legendary_counts.plot(kind='bar', color=['skyblue', 'orange'], figsize=(8, 6), rot=0)
```

```python
#PartC #6 - Categorize the pokemon by ‘Type 1’, and calculate the mean HP of each ‘Type.’ Visualize the Results. What type of pokemon has highest mean HP?

# Group by 'Type 1' and calculate the mean HP for each type
mean_hp_by_type = all_pokemon.groupby('Type 1')['HP'].mean().sort_values()
mean_hp_by_type

# Create a horinzontal bar plot
mean_hp_by_type.plot(kind='barh', color='skyblue', figsize=(10, 8), xlabel = "HP", title = "type vs hp")

print("Dragons have the highest avg HP")
```

```python
#PartC #7 - Query the max Attack of each Pokemon by Type. Visualize the data.

# Group by 'Type 1' and find the max Attack for each type
max_attack_by_type = all_pokemon.groupby('Type 1')['Attack'].max().sort_values()

# Create a bar plot
max_attack_by_type.plot(kind='barh', color='skyblue', figsize=(10, 8))
```

```python
#PartC #8 - Is there a correlation between HP and Defense? Visualize data to explain your stance.

# Scatter plot of HP vs. Defense
all_pokemon.plot.scatter(x='HP', y='Defense', color='skyblue', figsize=(10, 6))

# ignoring the outliers and large spread areas of the plot, 
# looking at only the concentrated datapoints shows us a 
# positive upward linear trend/correlation between HP and Defense.
```
