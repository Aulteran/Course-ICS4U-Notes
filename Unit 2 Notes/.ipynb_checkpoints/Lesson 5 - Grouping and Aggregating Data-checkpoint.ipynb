```python
import pandas as pd
```

```python
nba = pd.read_csv('nba_all_elo.csv')
```

```python
# Aggregate a DS using .sum() or .mean()
```

```python
# Get an array of a DS, with each unique term
nba['fran_id'].unique()
```

```python
group_fran = nba.groupby('fran_id', sort = False)
```

```python
group_fran # Not super useful on its own -> simply a DataFrame Object
```

```python
# Use groupby alongside aggregation
group_fran['pts'].sum()
```

```python
# Another example
## GOAL: Chicago Bulls -> their reacord each year (W vs L)
year_results = nba[nba['fran_id'] == 'Bulls'].groupby(['year_id', 'game_result'])  # Use multiple clauses in groupby
#1) Query rows with franchise as Bulls
#2) Group by Year, then add up results of each game per year
#3) Use aggregate to produce dataframe
```

```python
year_results['game_id'].count()  # Use aggregator on DF object to create new WHAT?
```

```python
# Using groupby -> How many G did the GSW W/L in 2015
GSW = nba[(nba['fran_id'] == 'Warriors') & (nba['year_id'] == 2015)].groupby('game_result')['game_id'].count()  # Group game result, then act on game_id
GSW
```

```python
GSW_playoffs = nba[(nba['fran_id'] == 'Warriors') & (nba['year_id'] == 2015)].groupby(['is_playoffs', 'game_result'])['game_id'].count() 
GSW_playoffs

# x[i].groupby([y]).count(), x = DF, y = what you're sorting by, i = what you're sorting
```

```python
# Want to know how many pts the New york Knicks scored each year
nyk_pts_by_year = nba[nba['fran_id'] == 'Knicks'].groupby('year_id')['pts'].sum()

nyk_pts_by_year
```

```python
# From array to Line Graph:
nyk_pts_by_year.plot()
```

```python
# Change TYPE of graph
nyk_pts_by_year.plot(kind = 'bar')
```

```python
# Last 10 years
nyk_pts_by_year.tail(10).plot(kind = 'bar')  # use tail for last 10 years; head for first 10 years
```

```python
# Create a scatter plot comparing the Knicks' point against their Opponents' points after 2000 from OG DF
knicks_vs_world = nba[(nba['fran_id'] == 'Knicks') & (nba['year_id'] > 2000)] # OMIT .groupby(['pts','opp_pts'])['year_id'].sum()
# When using scatter plot, DON'T use DS from Df; Use OG DF
knicks_vs_world.plot(kind = 'scatter', x = 'pts', y = 'opp_pts')  # Use x and y conditions
```

```python
# Miami Heat W/L Pie Chart
mia_2013 = nba[(nba['fran_id'] == 'Heat') & (nba['year_id'] == 2013)]['game_result'].value_counts()  # Query by game_result and count how many of the unique values (W, L) there are
mia_2013.plot(kind = 'pie')
```
