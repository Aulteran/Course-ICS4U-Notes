<h2>Querying DataSets - A Breakdown</h2>

```python
current_decade = nba[nba['year_id'] > 2010]  # DataFrame[condition] 

current_decade.shape
```

```python
nba['year_id'] > 2010
# Returns a True or False value depending on conidtion
# Query all 126314 rows from OG DF
# In this case, current_decade holds a new DF object
# In current_decade, return all rows where condition is true from the existing dataframe
```

<h1>Conditional Operators in Jupyter</h1>

+ AND = &
+ OR = |
+ NOT = ~
+ XOR = ^

```python
# Return an array of only the first occurence of each team id
nba['team_id'].unique()
```

```python
# Conditional operators and a query
nba[(nba['team_id'] == 'CLE') & (nba['pts'] > 100) & (nba['opp_pts'] > 100)]  # Parentheses allow each condition to be evaluated separately from one another
```
