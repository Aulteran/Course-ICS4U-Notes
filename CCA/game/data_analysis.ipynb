```python
'''
Author: Aadil Hussain
Built on: Python 3.11.7
'''

import pandas as pd
import matplotlib.pyplot

DATABASE_FILEPATH = 'database.csv'

game_data = pd.read_csv(DATABASE_FILEPATH)
```

```python
game_data.head()
```

```python
grouped_plant_zomb_correlation = game_data.groupby('NumPlants')['ZombiesKilled'].mean()
graph_title = "# of Plants Owned vs Avg Zombies Killed"
grouped_plant_zomb_correlation.plot(kind='line', x='NumPlants', y='ZombiesKilled', title=graph_title, ylabel='Avg # of Zombies Killed')
```
